```python
"""
내 정보를 담고 있는 신용카드를 여러개 만들려고 한다면?
"""
name = "Kim han kyung"
age = 28
addr = "seoul"
budget = 5000000
```

```python
def shinrak_card(name, age, addr, budget):
    print("-------------------------")
    print("신용카드 : 신락카드")
    print("이름: %s" % name)
    print("나이: %s" % age)
    print(f"신용한도: {budget}")
    print("-------------------------")

def kokbab_card(name, age, addr, budget):
    print("-------------------------")
    print("신용카드 : 국밥카드")
    print("이름: %s" % name)
    print("나이: %s" % age)
    print(f"신용한도: {budget}")
    print("-------------------------")

def dul_card(name, age, addr, budget):
    print("-------------------------")
    print("신용카드 : 둘카드")
    print("이름: %s" % name)
    print("나이: %s" % age)
    print(f"신용한도: {budget}")
    print("-------------------------")

def nuhe_card(name, age, addr, budget):
    print("-------------------------")
    print("신용카드 : 너희카드")
    print("이름: %s" % name)
    print("나이: %s" % age)
    print(f"신용한도: {budget}")
    print("-------------------------")
```

```python
shinrak_card(name, age, addr, budget)
kokbab_card(name, age, addr, budget)
dul_card(name, age, addr, budget)
nuhe_card(name, age, addr, budget)
```

```python
class my_card:
    def __init__(self, name, age, addr, budget):
        self.name = name
        self.age=age
        self.addr=addr
        self.budget=budget
```

```python
class my_card:
    def __init__(self, name, age, addr, budget):
        self.name = name
        self.age=age
        self.addr=addr
        self.budget=budget
        print("-------------------------")
        print("이름: %s" % name)
        print("나이: %s" % age)
        print(f"신용한도: {budget}")
        print("-------------------------")

class shinrak_card(my_card):
    def __init__(self, name, age, addr, budget):
        super().__init__(name, age, addr, budget)
        self.card_name = "신락카드"
        print(f"신용카드 : {self.card_name}")

class kokbab_card(my_card):
    def __init__(self, name, age, addr, budget):
        super().__init__(name, age, addr, budget)
        self.card_name = "국밥카드"
        print(f"신용카드 : {self.card_name}")

class dul_card(my_card):
    def __init__(self, name, age, addr, budget):
        super().__init__(name, age, addr, budget)
        self.card_name = "둘카드"
        print(f"신용카드 : {self.card_name}")

class nuhe_card(my_card):
    def __init__(self, name, age, addr, budget):
        super().__init__(name, age, addr, budget)
        self.card_name = "너희카드"
        print(f"신용카드 : {self.card_name}")
```
