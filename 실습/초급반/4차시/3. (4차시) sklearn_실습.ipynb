## Encoding

sklearn에서는 Label encoder, One-hot encoder, ordinal encoder을 제공

인코딩이라는 것은, str 형태로 된 데이터 친구들은, 사람은 읽을 수 있지만 컴퓨터는 읽을 수 없으니 이러한 데이터를 컴퓨터가 읽을 수 있게 숫자 값으로 변환해주는 과정

따라서 데이터가 숫자만으로 이루어진 게 아니라 범주형 변수 등으로 이루어져 있다면 인코딩을 사용해주시는 게 좋습니다.

### Label Encoder

````python
sample_data = pd.read_csv("./user_personalized_features.csv", index_col="Unnamed: 0")
````

````python
sample_data.head()
````

````python
from sklearn.preprocessing import LabelEncoder

data_copy = sample_data.copy()

# 라벨 인코더 엔진을 선언
encoder = LabelEncoder()

# 숫자형 컬럼이 아닌 것들을 라벨인코더로 변환 (fit)
data_copy['Gender'] = encoder.fit_transform(data_copy['Gender'])
data_copy['Location'] = encoder.fit_transform(data_copy['Location'])
data_copy['Interests'] = encoder.fit_transform(data_copy['Interests'])
data_copy['Product_Category_Preference'] = encoder.fit_transform(data_copy['Product_Category_Preference'])
````

### PCA

PCA는 특징(Feature)가 대량으로 구성되어있는 데이터셋을 저차원으로 축소시켜주는 차원축소 개념의 방법

기존의 데이터들을 특정 차원에 정사영했을 시 분산이 가장 큰 차원으로 축소시켜면서 축소시키고 싶은 N차원이 될 때까지 고유벡터와 고유값을 계산하여 계산된 고유벡터에 정사영 시키는 알고리즘

이로써 얻게되는 이점은

1. 노이즈 제거
2. 시각화 가능
3. 차원 축소
4. 메모리 효율화
5. 모델 성능

등이 있다.

````python
import pandas as pd
data_copy = sample_data.copy()

data_copy = pd.get_dummies(data_copy, columns=['Gender', 'Location', 'Interests', 'Product_Category_Preference'])
data_copy.head()
````

````python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

data_copy = data_copy.drop(['Newsletter_Subscription','User_ID'], axis=1)
x = data_copy.values 

# PCA를 선언하고 주 성분을 몇개로 줄일지 지정 
pca = PCA(n_components=2)

# PCA를 시킬 데이터를 지정하고 변환 
principalComponents = pca.fit_transform(x)

# PCA로 주성분분석이 끝난 데이터를 데이터프레임화 시키기
principalDf = pd.DataFrame(data=principalComponents, columns=['principal component 1', 'principal component 2'])

# 주성분마다 가지는 분산량 확인
print(pca.explained_variance_ratio_)
````

## Feature Scaling
서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업을 피처 스케일링(feature scaling)이라고 합니다. 대표적인 방법으로 표준화(Standardization) 와 정규화(Normalization) 이 있습니다.

표준화(Standardizaion)
표준화란 데이터의 피쳐 각각이 평균이 0이고 분산이 1인 가우시안 정규 분포를 가진 값으로 변환하는 것을 의미합니다.

정규화(Normalization)
정규화는 서로 다른 피처의 크기를 통일하기 위해 크기를 변환해주는 개념입니다.
```예를 들어 피처 A는 거리로서 0~100km 데이터이고, 피처 B는 금액으로서 0~100,000,000₩ 단위라면, 이 둘을 비교할 수 있는 수준의 단위로 만들어줘야 할 것입니다. 이를 정규화를 통해 진행합니다.```

````python
import matplotlib.pyplot as plt

# 데이터프레임 변수명 가져오기 
column_names = list(df.columns)

# 시각화 시킬 plot 즉, 캔버스 만들기 
fig, ax = plt.subplots(figsize=(15, 10))

# 변수마다 캔버스에 올리기
for i, column_name in enumerate(column_names):
    ax.hist(df[column_name], bins=10, alpha=0.5, label=column_name)

# 주석과 타이틀 지정 
ax.legend()
ax.set_title("Histograms of all variables in the DataFrame")

# 시각화 시작
plt.show()
````

````python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 정규화 시키는 class 선언 
scaler = StandardScaler()

# 정규화 시킬 데이터프레임 지정
scaler.fit(df)

# 해당 데이터프레임 정규화 변환
df_normalized = scaler.transform(df)

# 변환된 데이터 데이터프레임화 수행
df_normalized = pd.DataFrame(df_normalized, columns=column_names)
````

````python
import matplotlib.pyplot as plt

column_names = list(df_normalized.columns)
fig, ax = plt.subplots(figsize=(15, 10))

for i, column_name in enumerate(column_names):
    ax.hist(df_normalized[column_name], bins=10, alpha=0.5, label=column_name)

ax.legend()
ax.set_title("Histograms of all variables in the DataFrame")

plt.show()
````

````python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

for i, column_name in enumerate(column_names):
    ax1.hist(df[column_name], bins=10, alpha=0.5, label=column_name)

ax1.legend()
ax1.set_title("Histograms of all variables before Scaler")

for i, column_name in enumerate(column_names):
    ax2.hist(df_normalized[column_name], bins=10, alpha=0.5, label=column_name)

ax2.legend()
ax2.set_title("Histograms of all variables after Scaler")

plt.show()
````
