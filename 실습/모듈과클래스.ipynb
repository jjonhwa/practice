# 실습

### 실습 예

은행 계좌를 관리해주는 ATM기를 구현하려고 합니다. 클래스의 첫 인자는 계좌를 틀 사람명과 현재 잔액을 입력받을 수 있고 기능은 다음 3개를 구현하려고 합니다.

1. 입금 - 마이너스 통장이 아니기 때문에 입금은 양수만 가능하며, 입금 시 계좌의 잔고를 바로 파악할 수 있도록 구현

2. 출금 - 잔고에서 출금을 할 수 있는 기능을 구현하려 합니다. 출금 시 마이너스 통장이 아니기 때문에 잔고보다 클 수 없으며 마찬가지로 출금 금액도 양수여야 합니다.

3. 잔고확인 - 현재 잔고를 확인합니다.

class를 다음과 같이 작성할 수 있습니다.

```python
class BankAccount:
    def __init__(self, account_holder, balance=0.0):
        self.account_holder = account_holder
        self.balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount} 입금 완료. 현재 잔액은 다음과 같습니다. {self.balance}")
        else:
            print("입금은 양수만 가능합니다.")
    
    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"{amount} 인출 완료. 잔액은 다음과 같습니다. {self.balance}")
        else:
            print("출금 금액은 양수여야 하며 잔액보다 작거나 같아야 합니다.")
    
    def check_balance(self):
        print(f"현재 잔액은 다음과 같습니다. {self.balance}")
```

```python
# 실습 예시
alice = BankAccount("Alice")
alice.deposit(100)
alice.withdraw(50)
alice.check_balance()
```

```python
bruce = BankAccount("bruce", 100000)
bruce.check_balance()
```

```python
alice.check_balance()
```

```python
bruce.withdraw(10000000)
```
