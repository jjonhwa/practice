# 완전 기초

## Indexing

```python
sample_list = [5,1,2,3,6,8,9]
```

```python
'''
i번째 위치에 있는 데이터를 확인하고 싶을 때 Indexing을 한다라고 표현
실제로 코드상에서 구현은 다음과 같음
'''

sample_list[0]
```

```python
'''
데이터의 개수를 확인할 수 있는 함수 len()
'''

len(sample_list)
```

```python
'''
프로그램 언어는 0부터 시작이기 때문에 데이터의 개수가 7이면 마지막 숫자는 6
'''

sample_list[7]
```

```python
'''
위치를 조회할 땐 한개씩 안보고 여러개 조회가 가능
다음과 같은 코드로 수행하여 0,1,2번째 데이터를 한번에 조회할 수 있음
'''

sample_list[0:3]
```

```python
'''
데이터의 개수를 파악하기 어렵거나 헷갈릴때, 뒤에서부터 조회가 가능함
'''

sample_list[-1]
```

```python
'''
물론, 똑같이 여러 데이터를 조회할 수 있음
'''

sample_list[-3:-1]
```

```python
'''
시작이나 끝을 지정해주지 않으면 시작점부터 끝까지 조회
'''

sample_list[-3:]
```

## Other functions

```python
'''
데이터 중 가장 큰 수, 가장 작은 수를 기본적으로 확인할 수 있음
'''

print(f"가장 큰 수는 : {max(sample_list)}")
print(f"가장 작은 수는 : {min(sample_list)}")
```

```python
'''
데이터를 크기에 따라 나열할 수 있음
'''

print(f"오름차순 : {sorted(sample_list)}")
print(f"내림차순 : {sorted(sample_list, reverse=True)}")
```

```python
'''
데이터를 다 더한 값을 나타낼 수 있음
'''

print(f"데이터의 합은 : {sum(sample_list)}")
```

## List convert

```python
'''
리스트에 데이터를 추가하고 싶다!
다음과 같은 코드로 리스트의 "뒤에" 추가할 수 있음
'''

print(f"원래 데이터 리스트는 : {sample_list}")
copy_list = sample_list
copy_list.append(10)
print(f"10이 추가된 데이터 리스트는 : {copy_list}")
```

```python
'''
리스트에 데이터를 추가하고 싶다!
다음과 같은 코드로 리스트의 "뒤에" 추가할 수 있음
'''

print(f"원래 데이터 리스트는 : {sample_list}")
copy_list = sample_list
copy_list.append(10)
print(f"10이 추가된 데이터 리스트는 : {copy_list}")
```

```python
'''
앞에 추가하고 싶다!
'''

print(f"원래 데이터 리스트는 : {copy_list}")
copy_list.insert(0,0)
print(f"0이 추가된 데이터 리스트는 : {copy_list}")
```

```python
'''
다시 빼고싶다!
'''

print(f"원래 데이터 리스트는 : {copy_list}")
copy_list.pop(0)
print(f"0이 추가된 데이터 리스트는 : {copy_list}")
```
