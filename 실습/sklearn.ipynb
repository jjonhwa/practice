## Encoding

sklearn에서는 Label encoder, One-hot encoder, ordinal encoder을 제공

인코딩이라는 것은, str 형태로 된 데이터 친구들은, 사람은 읽을 수 있지만 컴퓨터는 읽을 수 없으니 이러한 데이터를 컴퓨터가 읽을 수 있게 숫자 값으로 변환해주는 과정

따라서 데이터가 숫자만으로 이루어진 게 아니라 범주형 변수 등으로 이루어져 있다면 인코딩을 사용해주시는 게 좋습니다.

### Label Encoder

````python
sample_data = pd.read_csv("/content/user_personalized_features.csv", index_col="Unnamed: 0")
````

````python
sample_data.head()
````

````python
from sklearn.preprocessing import LabelEncoder

data_copy = sample_data.copy()
# Create a LabelEncoder object
encoder = LabelEncoder()

# Fit the encoder on the categorical columns
data_copy['Gender'] = encoder.fit_transform(data_copy['Gender'])
data_copy['Location'] = encoder.fit_transform(data_copy['Location'])
data_copy['Interests'] = encoder.fit_transform(data_copy['Interests'])
data_copy['Product_Category_Preference'] = encoder.fit_transform(data_copy['Product_Category_Preference'])
````

### One-hot encoder

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder object
encoder = OneHotEncoder(sparse=False)

# Fit the encoder on the categorical columns
data_copy = pd.get_dummies(data_copy, columns=['Gender', 'Location', 'Interests', 'Product_Category_Preference'])

# Print the transformed data
print(data_copy.head())
````

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data_copy = sample_data.copy()

# Create a OneHotEncoder object
encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder on the categorical columns
data_copy = pd.DataFrame(encoder.fit_transform(data_copy[['Gender', 'Location', 'Interests', 'Product_Category_Preference']]))

# Set the column names for the encoded features
data_copy.columns = encoder.get_feature_names_out(['Gender', 'Location', 'Interests', 'Product_Category_Preference'])

# Print the encoded data
data_copy.head()
````

````python
import pandas as pd
data_copy = sample_data.copy()

# Fit the encoder on the categorical columns
data_copy = pd.get_dummies(data_copy, columns=['Gender', 'Location', 'Interests', 'Product_Category_Preference'])

# Print the transformed data
data_copy.head()
````

````python
# prompt: ordinal encoder 사용하는 코드

import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

data_copy = sample_data.copy()

# Create an OrdinalEncoder object
encoder = OrdinalEncoder()

# Fit the encoder on the categorical columns
data_copy = pd.DataFrame(encoder.fit_transform(data_copy[['Gender', 'Location', 'Interests', 'Product_Category_Preference']]))

# Set the column names for the encoded features
data_copy.columns = encoder.get_feature_names_out(['Gender', 'Location', 'Interests', 'Product_Category_Preference'])

# Print the encoded data
data_copy.head()
````

## Feature Scaling
서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업을 피처 스케일링(feature scaling)이라고 합니다. 대표적인 방법으로 표준화(Standardization) 와 정규화(Normalization) 이 있습니다.

표준화(Standardizaion)
표준화란 데이터의 피쳐 각각이 평균이 0이고 분산이 1인 가우시안 정규 분포를 가진 값으로 변환하는 것을 의미합니다.

정규화(Normalization)
정규화는 서로 다른 피처의 크기를 통일하기 위해 크기를 변환해주는 개념입니다.
```예를 들어 피처 A는 거리로서 0~100km 데이터이고, 피처 B는 금액으로서 0~100,000,000₩ 단위라면, 이 둘을 비교할 수 있는 수준의 단위로 만들어줘야 할 것입니다. 이를 정규화를 통해 진행합니다.```

````python
import matplotlib.pyplot as plt

# Create a list of column names
column_names = list(df.columns)

# Create a figure with a single subplot
fig, ax = plt.subplots(figsize=(15, 10))

# Iterate through each column and plot its histogram
for i, column_name in enumerate(column_names):
    ax.hist(df[column_name], bins=10, alpha=0.5, label=column_name)

# Add a legend and title
ax.legend()
ax.set_title("Histograms of all variables in the DataFrame")

# Show the plot
plt.show()
````

````python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the data in df
scaler.fit(df)

# Transform the data and store it in df_normalized
df_normalized = scaler.transform(df)

# Convert the numpy array back to a DataFrame
df_normalized = pd.DataFrame(df_normalized, columns=column_names)
````

````python
import matplotlib.pyplot as plt

# Create a list of column names
column_names = list(df_normalized.columns)

# Create a figure with a single subplot
fig, ax = plt.subplots(figsize=(15, 10))

# Iterate through each column and plot its histogram
for i, column_name in enumerate(column_names):
    ax.hist(df_normalized[column_name], bins=10, alpha=0.5, label=column_name)

# Add a legend and title
ax.legend()
ax.set_title("Histograms of all variables in the DataFrame")

# Show the plot
plt.show()

````

````python
import matplotlib.pyplot as plt

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot the histograms of the original data on the first subplot
for i, column_name in enumerate(column_names):
    ax1.hist(df[column_name], bins=10, alpha=0.5, label=column_name)

# Add a legend and title to the first subplot
ax1.legend()
ax1.set_title("Histograms of all variables before Scaler")

# Plot the histograms of the normalized data on the second subplot
for i, column_name in enumerate(column_names):
    ax2.hist(df_normalized[column_name], bins=10, alpha=0.5, label=column_name)

# Add a legend and title to the second subplot
ax2.legend()
ax2.set_title("Histograms of all variables after Scaler")

# Show the plot
plt.show()
````

````python
import pandas as pd
from sklearn.preprocessing import Normalizer

column_names = list(df.columns)

# Create a Normalizer object
normalizer = Normalizer()

# Fit the normalizer to the data in df
normalizer.fit(df)

# Transform the data and store it in df_normalized
df_normalized = normalizer.transform(df)

# Convert the numpy array back to a DataFrame
df_normalized = pd.DataFrame(df_normalized, columns=column_names)
````
